Load Dataset

```python
df = pd.read_csv("C:\\Users\\Asus\\Clone files\\Data_Driven_Water_Quality_Surveilance_And_Loss_Detection_System\\technical\\datasets\\water_quality_clean.csv")
print("Dataset Loaded Successfully")
df.head()
```

DATASET INFORMATION

```python
print("Dataset Shape:", df.shape)
print("\nDataset Columns:", df.columns)
print("\nData Types:", df.dtypes)
print("\nDataset Info:")
df.info()
```

STATISTICQL SUMMARY OF THE DATASET

```python
df.describe()
```

CHECK MISSING VALUES

```python
missing = df.isnull().sum()
missing_df = pd.DataFrame({'column_name': df.columns, 'missing_values': missing})   
print("\nMissing Values:")
print(missing_df)
```

```python
df['alert_flag'] = df['alert_flag'].fillna('normal')
```

```python
missing = df.isnull().sum()
missing_df = pd.DataFrame({'column_name': df.columns, 'missing_values': missing})   
print("\nMissing Values:")
print(missing_df)
```

CHECK DUPLICATES

```python
duplicates =df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicates)
```

CONVERT TIMESTEPS

```python
df['timestamp'] = pd.to_datetime(df['timestamp'],errors='coerce')
df['timestamp'].head()
```

Check Invalid Timestamps

```python
invalid_timestamp = df['timestamp'].isnull().sum()

print("Invalid timestamps:", invalid_timestamp)
```

Sensor Validation

```python
print("Unique Sensors:",
      df['sensor_id'].nunique())

print("Unique Locations:",
      df['location_id'].nunique())

print("Unique Sensor names:", df['sensor_id'].unique())
print("Unique Location names:", df['location_id'].unique())
```

Check Sensor Frequency

```python
sensor_frequency = (
    df.groupby('sensor_id')
      .size()
      .sort_values(ascending=False)
)

sensor_frequency.head()
```

Check Alert Flag

```python
print(
    df['alert_flag']
    .value_counts(dropna=False)
)
```

CONTAMINATION LABEL

```python
contamination_labels = [
    'wq_anomaly_tds_spike',
    'wq_anomaly_contamination_ph_drop',
    'wq_anomaly_turbidity_spike'
]

df['contamination_target'] = (
    df['alert_flag']
    .isin(contamination_labels)
    .astype(int)
)
```

LEAK LABEL

```python
df['leak_target'] = (
    df['alert_flag']
    == 'leak_suspected'
).astype(int)
```

```python
print("\nContamination Distribution:")
print(df['contamination_target'].value_counts())

print("\nLeak Distribution:")
print(df['leak_target'].value_counts())
```

```python
missing = df.isnull().sum()
missing_df = pd.DataFrame({'column_name': df.columns, 'missing_values': missing})   
print("\nMissing Values:")
print(missing_df)
```

Validate Water Quality Values

```python
print("pH Range:")
print(df['pH'].min(),
      df['pH'].max())

print("\nTurbidity Range:")
print(df['turbidity'].min(),
      df['turbidity'].max())

print("\nTDS Range:")
print(df['TDS'].min(),
      df['TDS'].max())

print("\nDO Range:")
print(df['DO'].min(),
      df['DO'].max())
```
